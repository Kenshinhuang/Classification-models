Load the data.

```python
import pandas as pd
df=pd.read_csv("fertility_diagnosis.csv",header=None)
df
```

The dataset looks cleaned already. Let's separate the predictor variables and the target variable.

```python
X=df.drop(columns=9, axis=0)
X
```

```python
from sklearn import preprocessing
names=X.columns
scaler=preprocessing.StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X, columns=names)
X.head()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
y=pd.DataFrame(df.iloc[:,9])
clf = ExtraTreesClassifier(n_estimators=50)
clf = clf.fit(X, y)
scores=pd.DataFrame(clf.feature_importances_).rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False).nlargest(5,"scores")
scores
```

```python
X=X.iloc[:,[1,8,0,6,5]]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
tree = RandomForestClassifier(max_depth=7, class_weight="balanced_subsample",random_state=0)
tree
```

```python
tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=tree.predict(X_test)
print("Accuracy on test data is: ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

True positive is 0. If we want to use this model to screen for altered semen, it is basically useless.

```python
len(df[df.iloc[:,9]=="O"])
```

The data set is imbalanced. Let's re-sample the dataset and see if that can improve sensitivity.

```python
df_O=df[df.iloc[:,9]=="O"]
df_O
```

We are going to replicate the positive cases 7 times. 

```python
df_O=pd.concat([df_O]*7,ignore_index=True)
df_O
```

Now, the positive cases and negative cases are less imbalanced. With the replication of positive cases, the positive cases are now 35.3% of the dataset. Concatenate the replicated positive instances with the negative instances in the original dataset.

```python
df_N=df[df.iloc[:,9]=="N"]
df=pd.concat([df_N,df_O],axis=0)
df
```

Build the model again.

```python
X=df.drop(columns=9, axis=0)
X
```

```python
from sklearn import preprocessing
names=X.columns
scaler=preprocessing.StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X, columns=names)
X.head()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
y=pd.DataFrame(df.iloc[:,9])
clf = ExtraTreesClassifier(n_estimators=50)
clf = clf.fit(X, y)
scores=pd.DataFrame(clf.feature_importances_).rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False).nlargest(5,"scores")
scores
```

```python
X=X.iloc[:,[1,8,0,6,5]]
X
```

```python
y=pd.DataFrame(df.iloc[:,9])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
tree = RandomForestClassifier(max_depth=10, class_weight="balanced_subsample",criterion="entropy",min_samples_split=4,random_state=0)
tree
```

```python
tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=tree.predict(X_test)
print("Accuracy on test data is: ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import roc_auc_score
print("Confusion matrix: \n",confusion_matrix(y_test,y_pred))
```

**By re-balancing the dataset, we went from a model with 87% overall accuracy on test data but 0% sensitivity to a model with 96% on test data, 100% sensitivity and 94.1% specificity.** Commendable results.

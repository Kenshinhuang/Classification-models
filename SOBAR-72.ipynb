```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_fa50cdf1f0d84f8f95a128ef173ba668.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.groupby("ca_cervix")["ca_cervix"].count()
```

```python
X=df.drop(["ca_cervix"],axis=1)
y=df[["ca_cervix"]]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
X.head()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
BForest=BalancedRandomForestClassifier(n_estimators=200,criterion="entropy",max_depth=6,min_samples_split=3,
                                       min_samples_leaf=2,max_features="sqrt",bootstrap=True,
                                       sampling_strategy="all",replacement=False,n_jobs=-1,warm_start=False,
                                       random_state=42)
scores=cross_val_score(BForest,X,np.asarray(y).reshape(y.shape[0],),cv=13)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
from sklearn.metrics import roc_auc_score
y_pred = cross_val_predict(BForest, X, np.asarray(y).reshape(y.shape[0],), cv=13)
AUC = roc_auc_score(y, y_pred)
print("AUC: ",AUC)
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
BBagging=BalancedBaggingClassifier(base_estimator=GaussianNB(),
                                    n_estimators=22,bootstrap_features=False,sampling_strategy="not majority",
                                   oob_score=False,n_jobs=-1,random_state=0)
scores=cross_val_score(BBagging,X,np.asarray(y).reshape(y.shape[0],),cv=8)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(BBagging, X, np.asarray(y).reshape(y.shape[0],), cv=8)
AUC = roc_auc_score(y, y_pred)
print("AUC: ",AUC)
```

```python
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
from imblearn.ensemble import EasyEnsembleClassifier
Ensemble=EasyEnsembleClassifier(base_estimator=GaussianNB(),n_estimators=20,sampling_strategy="not majority",random_state=0)
scores=cross_val_score(Ensemble,X,np.asarray(y).reshape(y.shape[0],),cv=8)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

Both BalancedBaggingClassifier and EasyEnsembleClassifier produced average accuracy of 93.06%. However,the BalanacedBaggingClassifier had a smaller standar deviation, implying that the BalancedBaggingClassifier is more stable than EasyEnsembleClassifier.

The BalancedBaggingClassifier model misclassified only 5 instances. The algorithm won't run on only 5 instances. Even if we just do a 2-fold cross validation, one of the folds would probably have only 1 class.(It has already been experimented.) Instead of running a reinforcement round on the instances that the model missed during the initial round, I will try to see if the model's performance can be further improved by adjusting the threshold.

```python
y_pred=pd.DataFrame(cross_val_predict(BBagging, X, np.asarray(y).reshape(y.shape[0],), 
                                      cv=8,method='predict_proba')).rename(columns={0:"pred_0",1:"pred_1"})
y_pred.head()
```

```python
y_probs1=y_pred[["pred_1"]]
```

```python
import numpy as np
thresholds=np.arange(0,1,0.001)
```

```python
def to_labels(pos_probs,threshold):
    return (pos_probs >= threshold).astype('int')
```

```python

from sklearn.metrics import f1_score
scores=[f1_score(y, to_labels(y_probs1,t),average="micro") for t in thresholds]
```

```python
ix=np.argmax(scores)
print((thresholds[ix],scores[ix]))
```

```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_820bb99c09934d7d9ebc132e45af625a.csv"
df=pd.read_csv(url,header=None)
df.head()
```

```python
df.shape
```

```python
def listToString(s):  
    str1 = " "   
    return (str1.join(s)) 
```

```python
dna=pd.DataFrame(df.iloc[:,2])
```

```python
import numpy as np
for i in np.arange(0,106):
    string=listToString(dna.loc[i].tolist())
    print(string)
```

```python
import numpy as np
dinucleotides = ['aa','at','ag','ac',
                 'ta','tt','tg','tc',
                 'ga','gt','gg','gc',
                 'ca','ct','cg','cc']
all_counts = []
for i in np.arange(0,106):
    string=listToString(dna.loc[i].tolist())
    for dinucleotide in dinucleotides:
        count = string.count(dinucleotide)
        print("count is " + str(count) + " for " + dinucleotide)
        all_counts.append(count)
print(all_counts)
```

```python
all_counts=np.asarray(all_counts).reshape(106,16)
all_counts
```

```python
X=pd.DataFrame(all_counts)
X
```

```python
df=df.rename(columns={0:"Class"})
y=df[["Class"]]
y.groupby("Class")["Class"].count()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X))
X
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
from sklearn import svm
model1=svm.SVC(random_state=0)
scores=cross_val_score(model1,X,np.asarray(y).reshape(y.shape[0],),cv=5)
print("Accuracy is: ",scores.mean(), "+/-",scores.std())
```

```python
y_pred = cross_val_predict(model1, X, np.asarray(y).reshape(y.shape[0],), cv=5)
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("Classification report: \n",classification_report(y,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier(n_estimators=100,random_state=0)
scores=cross_val_score(model2,X,np.asarray(y).reshape(y.shape[0],),cv=5)
print("Accuracy is: ",scores.mean(), "+/-",scores.std())
```

```python
y_pred = cross_val_predict(model2, X, np.asarray(y).reshape(y.shape[0],), cv=5)
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("Classification report: \n",classification_report(y,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model3=GradientBoostingClassifier(n_estimators=150,max_features="sqrt",random_state=0)
scores=cross_val_score(model3,X,np.asarray(y).reshape(y.shape[0],),cv=5)
print("Accuracy is: ",scores.mean(), "+/-",scores.std())
```

```python
from sklearn.neighbors import KNeighborsClassifier
model4=KNeighborsClassifier(n_neighbors=3)
scores=cross_val_score(model4,X,np.asarray(y).reshape(y.shape[0],),cv=4)
print("Accuracy is: ",scores.mean(), "+/-",scores.std())
```

```python
from sklearn.naive_bayes import GaussianNB
model5=GaussianNB()
scores=cross_val_score(model5,X,np.asarray(y).reshape(y.shape[0],),cv=5)
print("Accuracy is: ",scores.mean(), "+/-",scores.std())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
model6=svm.SVC(random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
model6
```

```python
from sklearn.metrics import accuracy_score
train_pred=model6.predict(X_train)
test_pred=model6.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
classes = df.iloc[:,0]
print(classes[:5])
```

```python
# generate list of DNA sequences
sequences = list(df.iloc[:, 2])
dataset = {}

# loop through sequences and split into individual nucleotides
for i, seq in enumerate(sequences):
    
    # split into nucleotides, remove tab characters
    nucleotides = list(seq)
    nucleotides = [x for x in nucleotides if x != '\t']
    
    # append class assignment
    nucleotides.append(classes[i])
    
    # add to dataset
    dataset[i] = nucleotides
    
print(dataset[0])
```

```python
X = pd.DataFrame(dataset)
X
```

```python
X=X.transpose()
X
```

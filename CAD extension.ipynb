```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_aa04e98df66b49989ab49f9f2ac12237.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
y=df[["LAD"]]
y.groupby("LAD")["LAD"].count()
```

```python
X=df.drop(["LAD","LCX","RCA","Cath"],axis=1)
X.head()
```

```python
types=pd.DataFrame(X.dtypes,columns=["type"])
types
```

```python
objects=types[types["type"]=="object"]
objects
```

```python
object_index=objects.index.tolist()
object_index
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
objects=X[object_index]
objects=objects.apply(le.fit_transform)
objects.head()
```

```python
numeric=types[types["type"]!="object"]
numeric
```

```python
numeric_index=numeric.index.tolist()
numeric_index
```

```python
numeric=X[numeric_index]
numeric.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numeric=pd.DataFrame(scaler.fit_transform(numeric),columns=numeric.columns)
numeric.head()
```

```python
X=pd.concat([objects,numeric],axis=1)
X.head()
```

```python
le2=LabelEncoder()
y=pd.DataFrame(le.fit_transform(y),columns=y.columns)
y.head()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
scores=scores.fillna(0)
```

```python
X_scores=scores[scores["scores"]>6]
len(X_scores)
```

```python
Xp=X.iloc[:,X_scores.index]
Xp.head()
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
from sklearn import svm
model1=svm.SVC(kernel="sigmoid",gamma="auto",random_state=0)
scores=cross_val_score(model1,Xp,np.asarray(y).reshape(y.shape[0],),cv=7)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(model1, Xp,np.asarray(y).reshape(y.shape[0],) , cv=7)
AUC = roc_auc_score(y, y_pred)
print("AUC: ",AUC)
```

```python
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
y.groupby("LAD")["LAD"].count()
```

```python
print("Specificity: ",89/126)
print("Sensitivity: ",150/177)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2=RandomForestClassifier(n_estimators=200,criterion="entropy",random_state=0)
scores=cross_val_score(model2,Xp,np.asarray(y).reshape(y.shape[0],),cv=7)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(model2, Xp,np.asarray(y).reshape(y.shape[0],) , cv=7)
AUC = roc_auc_score(y, y_pred)
print("AUC: ",AUC)
```

```python
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("Specificity: ",85/126)
print("Sensitivity: ",155/177)
```

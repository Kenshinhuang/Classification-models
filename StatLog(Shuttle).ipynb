```python
import pandas as pd
url2="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_aa4ff181881749dd8669f34e82c87c01.csv"
df2=pd.read_csv(url2,header=None)
df2.head()
```

```python
df2=df2.rename(columns={9:"Class"})
df2.head()
```

```python
X_test=df2.drop(["Class"],axis=1)
y_test=df2[["Class"]]
```

```python
y_test.groupby("Class")["Class"].count()
```

```python
y_test.groupby("Class")["Class"].count()/len(y_test)
```

```python
url1="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_9d91340770aa4a1483cef85ef184b206.csv"
df1=pd.read_csv(url1,header=None)
df1.head()
```

```python
df1.shape
```

```python
df1=df1.rename(columns={9:"Class"})
df1.head()
```

```python
X_train=df1.drop(["Class"],axis=1)
y_train=df1[["Class"]]
```

```python
y_train.groupby("Class")["Class"].count()
```

```python
y_train.groupby("Class")["Class"].count()/len(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=pd.DataFrame(scaler.fit_transform(X_train))
X_train.head()
```

```python
X_test=pd.DataFrame(scaler.fit_transform(X_test))
X_test.head()
```

```python
from sklearn.multiclass import OneVsOneClassifier
from sklearn import svm
import numpy as np

model=OneVsOneClassifier(svm.SVC(C=8,gamma="auto",random_state=0)).fit(X_train, np.asarray(y_train).reshape(y_train.shape[0],))
model
```

```python
from sklearn.metrics import accuracy_score
train_pred=model.predict(X_train)
test_pred=model.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
from sklearn.metrics import f1_score
print("F1 socre for train data: ",f1_score(y_train,train_pred,average="micro"))
print("F1 score for test data: ",f1_score(y_test,test_pred,average="micro"))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix for train data: \n",confusion_matrix(y_train,train_pred))
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model=OneVsOneClassifier(RandomForestClassifier(random_state=0)).fit(X_train, np.asarray(y_train).reshape(y_train.shape[0],))
model
```

```python
train_pred=model.predict(X_train)
test_pred=model.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
print("F1 socre for train data: ",f1_score(y_train,train_pred,average="micro"))
print("F1 score for test data: ",f1_score(y_test,test_pred,average="micro"))
```

```python
print("Confusion matrix for train data: \n",confusion_matrix(y_train,train_pred))
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model=OneVsOneClassifier(GradientBoostingClassifier(random_state=0)).fit(X_train, np.asarray(y_train).reshape(y_train.shape[0],))
model
```

```python
from sklearn.metrics import accuracy_score
train_pred=model.predict(X_train)
test_pred=model.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
print("F1 socre for train data: ",f1_score(y_train,train_pred,average="micro"))
print("F1 score for test data: ",f1_score(y_test,test_pred,average="micro"))
```

```python
print("Confusion matrix for train data: \n",confusion_matrix(y_train,train_pred))
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

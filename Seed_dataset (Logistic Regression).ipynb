```python
import pandas as pd
df=pd.read_csv("seeds_dataset.csv")
df.head()
```

```python
X=df.drop(["Varieties"],axis=1)
X.head()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
X=pd.DataFrame(est.fit_transform(X),columns=X.columns)
X.head()
```

```python
y=df[["Varieties"]]
y.groupby("Varieties")["Varieties"].count()
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
import numpy as np
```

```python
LR=LogisticRegression(C=3,solver="lbfgs",random_state=0)
scores=cross_val_score(LR,X,np.asarray(y).reshape(y.shape[0],),cv=10)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(LR, X,np.asarray(y).reshape(y.shape[0],), cv=10)
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
from sklearn.metrics import classification_report
print("Classification report: \n",classification_report(y,y_pred))
```

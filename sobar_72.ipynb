```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_5aa6bf17984840e9901062d8dce3c4cd.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
X=df.drop(["ca_cervix"],axis=1)
X.head()
```

```python
y=df[["ca_cervix"]]
y.value_counts()
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import LeaveOneOut
from sklearn import metrics
from sklearn.metrics import auc
from sklearn.metrics import plot_roc_curve
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
X_scores=scores[scores["scores"]>12]
len(X_scores)
```

```python
Xp=X.iloc[:,X_scores.index]
Xp.head()
```

```python
from sklearn import svm
model1=svm.SVC(random_state=0)
scores=cross_val_score(model1,Xp,np.asarray(y).reshape(y.shape[0],),cv=5)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(model1, Xp, np.asarray(y).reshape(y.shape[0],), cv=5)
print("AUC: ",roc_auc_score(y,y_pred))
```

```python
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("True positive rate: ",18/21)
print("True negative rate: ",49/51)
```

Load the data.

```python
import pandas as pd
df=pd.read_csv("advertisement.csv",header=None, low_memory=False)
df
```

The data has been cleaned on Excel. There are many missing values in columns 0, 1,,2,3. I replaced the missing values on columns 0, 1,2 with the column mean and replaced the missing values in column 3 with the mode. Python's "mean" and "mode" methods could not handle missing values. However, Excel's "AVERAGE" and "MODE" functions automatically ignore any non-numeric values and calculate the mean and mode, respectively, on numeric values. 

It is also a lot quicker to replace the question marks on Excel also. I used nested the "ISNUMBER" function inside the logical "IF" function and just clicked to fill the entire columns. 

Here is another reason to clean the data on Excel. "df.iloc[:,0]==" ?" returned FALSE for all rows, with or without the space in front of the question mark. All in all, for this dataset, it is a lot quicker to clean it on Excel and upload the cleaned dataset to the Python environment.

Separate the potential predictors from the target.

```python
X=df.drop(columns=1558,axis=1)
X
```

```python
y=pd.DataFrame(df.iloc[:,1558])
y
```

Feature selection.

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

clf = ExtraTreesClassifier(n_estimators=50)
clf = clf.fit(X, y)
scores=pd.DataFrame(clf.feature_importances_).rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False).nlargest(50,"scores")
scores
```

Extract the column indices of the selected features.

```python
import numpy as np
np.asarray(scores.index)
```

Form the predictor set.

```python
X=X.iloc[:,[1, 1399, 1243, 1229,  351, 1483,  968, 1435,    2,    0, 1047,
       1394,  874, 1108, 1153,  398, 1278, 1276, 1455, 1247, 1362, 1116,
       1344, 1385, 1424,  983, 1143,  860, 1022, 1040, 1445,  307,    3,
        317, 1309, 1218,  167, 1118,  187,  508, 1271, 1373, 1207, 1293,
       1198, 1254, 1326,  175, 1179, 1359]]
X
```

One-hot encode the target variable.

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
y=pd.DataFrame(enc.fit_transform(y).toarray())
y.head()
```

Split the data for training and testing.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

Build the neural network model.

```python
length=X_train.shape[1]
num_classes=y_test.shape[1]

import keras
from keras.models import Sequential
from keras.layers import Dense

# define classification model

def classification_model():
    # create model
    model = Sequential()
    model.add(Dense(length, activation='relu', input_shape=(length,)))
    model.add(Dense(65,activation='relu'))
    model.add(Dense(65,activation="relu"))
    model.add(Dense(65,activation="relu"))
    model.add(Dense(65,activation="relu"))
    model.add(Dense(65,activation="tanh"))
    model.add(Dense(num_classes, activation='softmax'))
    
    
    # compile model
    model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

Train the model on train data and test it on test data.

```python
model=classification_model()
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20, verbose=2)
model.evaluate(X_test,y_test,verbose=0)
```

It is interesting that I experimented using 50, 100 or 200 predictor features and there was no significant difference on final accuracy on test data. In all cases, the final accuracy on test data ranged rom 95.5% to 96.5%.

```python
from sklearn.metrics import roc_auc_score
print("AUC: ",roc_auc_score(y_test,model.predict(X_test)))
```

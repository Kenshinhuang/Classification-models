```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_0cc49a4247064ece92a2b2ec1f62d4e9.csv"
df=pd.read_csv(url,header=None)
df.head()
```

```python
df.shape
```

```python
df=df.rename(columns={0:"Class"})
y=df[["Class"]]
y.groupby("Class")["Class"].count()
```

```python
classes = df.iloc[:,0]
print(classes[:5])
```

```python
sequences = list(df.iloc[:, 2])
dataset = {}

for i, seq in enumerate(sequences):
    nucleotides = list(seq)
    nucleotides = [x for x in nucleotides if x != '\t']
    nucleotides.append(classes[i])
    dataset[i] = nucleotides
    
print(dataset[0])
```

```python
X = pd.DataFrame(dataset)
X
```

```python
X=X.transpose()
X.head()
```

```python
X=X.iloc[:,0:60]
X.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(handle_unknown="ignore")
X=pd.DataFrame(enc.fit_transform(X).toarray())
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn import svm
import numpy as np
model1=svm.SVC(kernel="poly",C=2,
               random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
model1
```

```python
from sklearn.metrics import accuracy_score
train_pred=model1.predict(X_train)
test_pred=model1.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

```python
from sklearn.multiclass import OneVsRestClassifier
model2 = OneVsRestClassifier(svm.SVC(kernel="poly",C=2,
                             random_state=0)).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
model2
```

```python
train_pred=model2.predict(X_train)
test_pred=model2.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

```python
from sklearn.multiclass import OneVsOneClassifier
model3 = OneVsOneClassifier(svm.SVC(kernel="poly",C=2,
                             random_state=0)).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
model3
```

```python
train_pred=model3.predict(X_train)
test_pred=model3.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

#### Summary:
- SVM with "poly" kernel: 97.11% accuracy on test data.
- SVM with "poly" kernel wrapped inside OneVsRestClassifier: 97.24% accuracy on test data.
- SVM with "poly" kernel wrapped inside OneVsOneClassifier: 97.11% accuracy on test data.

```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_7aa48789c7014f2ea5c0f4a07b76e4c2.csv"
df=pd.read_csv(url,header=None,na_values=["?"])
df.head()
```

```python
df.shape
```

```python
df.isna().any()
```

```python
df1=df.dropna(axis=1,thresh=0.8*df.shape[0])
```

```python
df.shape
```

```python
df=df.fillna(df.mean())
```

```python
df.isna().any()
```

```python
df=df.rename(columns={0:"Class"})
df.head()
```

```python
y=df[["Class"]]
y.value_counts()
```

```python
X=df.drop(["Class"],axis=1)
X.head()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
X_scores=scores[scores["scores"]>5]
len(X_scores)
```

```python
Xp=X.iloc[:,X_scores.index]
Xp.head()
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
```

```python
from sklearn import svm
model1=svm.SVC(C=2,random_state=0)
scores=cross_val_score(model1,Xp,np.asarray(y).reshape(y.shape[0],),cv=3)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
from sklearn.multiclass import OneVsOneClassifier
model2= OneVsOneClassifier(svm.SVC(C=3,random_state=0))
scores=cross_val_score(model2,Xp,np.asarray(y).reshape(y.shape[0],),cv=3)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier(n_estimators=300,bootstrap=False,random_state=0)
scores=cross_val_score(model3,Xp,np.asarray(y).reshape(y.shape[0],),cv=3)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
from sklearn.multiclass import OneVsRestClassifier
model4= OneVsRestClassifier(svm.SVC(C=3,random_state=0))
scores=cross_val_score(model4,Xp,np.asarray(y).reshape(y.shape[0],),cv=3)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(model1, Xp, np.asarray(y).reshape(y.shape[0],), cv=3)
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

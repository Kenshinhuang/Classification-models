0 = Hypothyroid; 1 = Negative

```python
X=df.drop(["Class"],axis=1)
X.head()
```

```python
y=df[["Class"]]
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X))
X.head()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
X_scores=scores[scores["scores"]>=1]
len(X_scores)
```

```python
X=X.iloc[:,X_scores.index]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
y_train.groupby("Class")["Class"].count()
```

```python
y_test.groupby("Class")["Class"].count()
```

```python
print("Percentage of majority class in train data: ",2014/2119)
print("Percentage of majority class in test data: ",998/1044)
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
import numpy as np
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
BForest=BalancedRandomForestClassifier(n_estimators=200,criterion="entropy",max_depth=5,min_samples_split=4,
                                       min_samples_leaf=2,max_features="log2",bootstrap=False,
                                       sampling_strategy="not majority",replacement=True,n_jobs=-1,warm_start=False,
                                       random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
BForest
```

```python
from sklearn.metrics import accuracy_score
train_pred=BForest.predict(X_train)
test_pred=BForest.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

Majority class in both train and test data is about 95%. Hence, for the model to have any significance, accuracy must be above 95% on both train and test data. This is a dataset with highly imbalanced class distribution.

Let's try another classification algorithm and see if we could get past 98.63% accuracy on train data and 97.7% accuracy on test data.

```python
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
BBagging=BalancedBaggingClassifier(base_estimator=GradientBoostingClassifier(n_estimators=150),
                                    n_estimators=30,bootstrap_features=True,sampling_strategy="not majority",
                                    n_jobs=-1,random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
BBagging
```

```python
train_pred=BBagging.predict(X_train)
test_pred=BBagging.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
BBagging_svm=BalancedBaggingClassifier(base_estimator=svm.SVC(kernel="linear"),
                                    n_estimators=30,bootstrap_features=True,sampling_strategy="not majority",
                                    n_jobs=-1,random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
BBagging_svm
```

```python
test_pred=BBagging_svm.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

Since there is only 1 class in the missed instances from round 1's train data. I cannot just re-train the model on the missed instances from the inital train data and test it on the missed instances from the initial test data. There must be at least 2 classes. Hence, the missed instances from the initial round's train and test data must be concatenated and split again. 

```python
round2=pd.concat([train1,test1],axis=0)
round2.groupby("Class")["Class"].count()
```

```python
X1=round2.drop(["Class"],axis=1)
y1=round2[["Class"]]
```

```python
X_train1,X_test1,y_train1,y_test1=train_test_split(X1,y1,test_size=0.62,random_state=42)
print(X_train1.shape)
print(X_test1.shape)
```

```python
BBagging1=BalancedBaggingClassifier(base_estimator=GradientBoostingClassifier(n_estimators=200,max_features="auto"),
                                    n_estimators=35,bootstrap_features=False,sampling_strategy="all",
                                    n_jobs=-1,random_state=0).fit(X_train1,np.asarray(y_train1).reshape(y_train1.shape[0],))
BBagging1
```

```python
train_pred1=BBagging1.predict(X_train1)
test_pred1=BBagging1.predict(X_test1)
print("Accuracy on train data for the 2nd round: ",accuracy_score(y_train1,train_pred1))
print("Accuracy on test data for the 2nd round: ",accuracy_score(y_test1,test_pred1))
```

```python
print("Confusion on train data for the 2nd round: \n",confusion_matrix(y_train1,train_pred1))
print("Confusion on test data for the 2nd round: \n",confusion_matrix(y_test1,test_pred1))
```

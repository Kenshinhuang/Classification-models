```python
import pandas as pd
url1="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_00b6f72ab80c4b98a1827b0ff932dd4e.csv"
df1=pd.read_csv(url1)
df1.head()
```

```python
df1.shape
```

```python
X=df1.drop(["Unnamed: 0"],axis=1)
X.head()
```

```python
url2="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_e33d8807b7cf426ca530088d33640b5e.csv"
df2=pd.read_csv(url2)
df2.head()
```

```python
y=df2.drop(["Unnamed: 0"],axis=1)
y.head()
```

```python

import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
scores=scores.fillna(0)
scores
```

```python
scores=scores[scores["scores"]>0]
scores
```

```python
import matplotlib.pyplot as plt
plt.hist(scores)
plt.show()
```

A handful of the features have scores that are much higher than the rest of the features, i.e., above 1,500. Then, there are about 16,000 features with scores at or below 330. This means of many of the features are insignificant. Including these insignificant features in the model will not only slow down the model (there are 16,383 features in total), but also add noise to the model and compromise the model's performance.

```python
X_scores=scores[scores["scores"]>1900]
len(X_scores)
```

```python
Xp=X.iloc[:,X_scores.index]
Xp.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Xp=pd.DataFrame(scaler.fit_transform(Xp),columns=Xp.columns)
Xp.head()
```

```python
y.groupby("Class")["Class"].count()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xp,y,test_size=0.3,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn import svm
model1=svm.SVC(random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
model1
```

```python
from sklearn.metrics import accuracy_score
train_pred=model1.predict(X_train)
test_pred=model1.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix for train data: \n",confusion_matrix(y_train,train_pred))
print("Confusion matrix for test data: \n",confusion_matrix(y_test,test_pred))
```

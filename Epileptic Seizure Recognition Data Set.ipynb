```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_96e60c86a00b486ea5130e16275c1674.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
X=df.drop(["X0","y"],axis=1)
X.head()
```

```python
y=df[["y"]].rename(columns={"y":"Class"})
y.head()
```

```python
import numpy as np
np.unique(y)
```

```python
y=y.replace([2,3,4,5],0)
np.unique(y)
```

```python
y.groupby("Class")["Class"].count()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
X_scores=scores[scores["scores"]>8]
len(X_scores)
```

```python
X_prime=X.iloc[:,X_scores.index]
X_prime.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_prime=pd.DataFrame(scaler.fit_transform(X_prime),columns=X_prime.columns)
X_prime.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_prime,y,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
BForest=BalancedRandomForestClassifier(n_estimators=170,criterion="entropy",min_samples_split=3,min_samples_leaf=2,
                                       sampling_strategy="not majority",n_jobs=-1,
                                       random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
BForest
```

```python
from sklearn.metrics import accuracy_score
train_pred=BForest.predict(X_train)
test_pred=BForest.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
from sklearn.metrics import f1_score
print("F1 score on train data: ",f1_score(y_train,train_pred,average="micro"))
print("F1 score on test data: ",f1_score(y_test,test_pred,average="micro"))
```

```python
from sklearn.metrics import precision_score
print("Precision score on train data: ",precision_score(y_train,train_pred,average="micro"))
print("Precision score on test data: ",precision_score(y_test,test_pred,average="micro"))
```

```python
from sklearn.metrics import recall_score
print("Recall score on train data: ",recall_score(y_train,train_pred,average="micro"))
print("Recall score on test data: ",recall_score(y_test,test_pred,average="micro"))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix on train data: \n",confusion_matrix(y_train,train_pred))
print("Confusion matrix on test data: \n",confusion_matrix(y_test,test_pred))
```

The best performance on test data is delivered by the Support Vector Machine model when the minority class in the train data is duplicated by 2 times to increase the model's exposure to the minority class. With a slight tuning of just two of the hyper-parameters, the SVM model deliver a 97.913% accuracy on test data. The second best performance on test data is delivered by the SVM model as well, with no data resampling. In this case, by fine-tuning just 1 hyper-parameter, the SVM model delivered a 97.652% accuracy on test data. So, duplicating the minority class in train data by 2 times increased the best model's performance further by 0.261%. That number by itself does not sound like a lot. However, it we put it in the context of a dataset with more than 10,000 rows of data, an increase of 0.261% in test data accuracy means that the model is able to make **2,610** more correct predictions. 

Besides the quick data resampling technique that I showed above, feature selection can also be an effective strategy to tackle class imbalance and improve model performance. The entire dataset has 178 potential predictor variables, I used only 53 of them. That is less than 30%. I implemented the SelectKBest with the "f_classif" function and set the threshold for feature score at 8. I found that lowring the feature score threshold further did not contribute to any improvement in model performance. On the contrary, including more less relevant features compromised even the best model's performance.

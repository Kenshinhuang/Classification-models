Load the **train data** into Pandas dataframe.

```python
import pandas as pd
df=pd.read_csv("KDD_census_income.csv",header=None)
df.head()
```

```python
df.shape
```

The train data has 199,523 rows and 42 columns.

Replace all missing categorical values that are currently denoted by "?" with the word "unknown".

```python
df=df.replace(" ?", "unknown")
df
```

A lot of the time, missing numeric values will are denoted by large numbers such as 9999, 99999, etc. They are usually the largest numbers in any numeric column. We don't really need the summary statistics, but the "describe" method can reveal the columns with such values.

```python
df.describe()
```

Columns 5, 16,18 have missing values and the missing values are being represented by large numbers such as 9999, 99999.

```python
df_5=pd.DataFrame(df.iloc[:,5])
df_5[df_5.loc[:,5]==9999]
mean_5=df_5.drop(index=190936,axis=0).mean()
df_5=df_5.replace(9999,mean_5)
```

```python
df_16=pd.DataFrame(df.iloc[:,16])
missing=df_16[df_16.loc[:,16]==99999]
index=missing.index
mean_16=df_16.drop(index=index,axis=0).mean()
df_16=df_16.replace(99999,mean_16)
```

```python
df_18=pd.DataFrame(df.iloc[:,18])
missing=df_18[df_18.loc[:,18]==99999]
index=missing.index
mean_18=df_18.drop(index=index,axis=0).mean()
df_18=df_18.replace(99999,mean_18)
```

The cleaned dataset:

```python
import numpy as np
df=df.drop(columns=[5,16,18],axis=1)
df=pd.concat([df_5,df_16,df_18,df],axis=1)
columns=np.arange(0,42)
df=df.reindex(columns=columns)
df.head()
```

```python
X=df.drop(columns=41,axis=1)
X.head()
```

Use the "describe" method to get the column indices of the numeric columns.

```python
X.describe().columns
```

Separate the numeric columns from the categorical columns.

```python
X_numeric=X.iloc[:,[0, 2, 3, 5, 16, 17, 18, 24, 30, 36, 38, 39, 40]]
X_numeric
```

```python
X_categorical=X.drop(columns=[0, 2, 3, 5, 16, 17, 18, 24, 30, 36, 38, 39, 40],axis=1)
```

```python
X_categorical
```

Label encode the categorical variables. 

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
X_categorical=X_categorical.apply(le.fit_transform)
X_categorical
```

Concatenate the encoded categorical variables and the numeric variables together.

```python
X=pd.concat([X_numeric,X_categorical],axis=1)
X.head()
```

Standardize the predictor variables.

```python
from sklearn import preprocessing
names=X.columns
scaler=preprocessing.StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X, columns=names)
X.head()
```

Get the target variable.

```python
y=pd.DataFrame(df.loc[:,41]).rename(columns={41:"income"})
y.head()
```

Feature importance scores for all features.

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
clf = ExtraTreesClassifier(n_estimators=50)
clf = clf.fit(X, y)
scores=clf.feature_importances_ 
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

Select the top 20 features.

```python
top_20=scores.nlargest(20,"scores")
top_20.index
```

Form the predicto set with the top 20 features.

```python
X=X.iloc[:,top_20.index]
X.head()
```

```python
df=pd.concat([X,y],axis=1)
df
```

Check whether classes are evenly distributed.

```python
y.groupby(["income"])["income"].count()
```

Un-even class distribution. Re-calibration is needed. We are going to over-sample the minority class "50000+."

```python
df_above=df[df["income"]==" 50000+."]
df_above
```

```python
df_above=pd.concat([df_above]*14,axis=0)
df_above
```

Concatenate the majority class and the over-sampled minority class together.

```python
df_below=df[df["income"]=="-50000"]
df=pd.concat([df_above,df_below],axis=0)
df
```

```python
X=df.drop(["income"],axis=1)
X.head()
```

```python
y=df[["income"]]
y.head()
```

Build the Random Forest model.

```python
from sklearn.ensemble import RandomForestClassifier
Tree = RandomForestClassifier(n_estimators=750,max_depth=34, random_state=0,criterion="entropy",max_features="sqrt")
Tree
```

```python
Tree.fit(X,y)
```

```python
from sklearn.metrics import accuracy_score
y_pred=Tree.predict(X)
print("Accuracy on train data: ",accuracy_score(y,y_pred))
```

Load **KDD's test data**.

```python
df_test=pd.read_csv("KDD_census_income_test.csv",header=None)
df_test
```

Clean the test data the way the train data was cleaned.

```python
df_test=df_test.replace(" ?", "unknown")
```

```python
df_test.describe()
```

```python
df_test_16=pd.DataFrame(df_test.iloc[:,16])
missing=df_test_16[df_test_16.loc[:,16]==99999]
index=missing.index
mean_16=df_test_16.drop(index=index,axis=0).mean()
df_test_16=df_test_16.replace(99999,mean_16)
```

```python
df_test_18=pd.DataFrame(df_test.iloc[:,18])
missing=df_test_18[df_test_18.loc[:,18]==99999]
index=missing.index
mean_18=df_test_18.drop(index=index,axis=0).mean()
df_test_18=df_test_18.replace(99999,mean_18)
```

```python
import numpy as np
df_test=df_test.drop(columns=[16,18],axis=1)
df_test=pd.concat([df_test_16,df_test_18,df_test],axis=1)
columns=np.arange(0,42)
df_test=df_test.reindex(columns=columns)
df_test.head()
```

```python
X_test=df_test.drop(columns=41,axis=1)
```

```python
X_test_numeric=X_test.iloc[:,[0, 2, 3, 5, 16, 17, 18, 24, 30, 36, 38, 39, 40]]
X_test_numeric.head()
```

```python
X_test_categorical=X_test.drop(columns=[0, 2, 3, 5, 16, 17, 18, 24, 30, 36, 38, 39, 40],axis=1)
```

```python
X_test_categorical=X_test_categorical.apply(le.fit_transform)
X_test_categorical
```

```python
X_test=pd.concat([X_test_numeric,X_test_categorical],axis=1)
```

```python
X_test=scaler.fit_transform(X_test)
X_test=pd.DataFrame(X_test)
X_test.head()
```

```python
X_test=X_test.iloc[:,top_20.index]
```

```python
X_test.shape
```

```python
y_test=pd.DataFrame(df_test.iloc[:,41])
y_test
```

Evaluate the model on the test data.

```python
y_test_pred=Tree.predict(X_test)
print("Accuracy on test data: ",accuracy_score(y_test,y_test_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix on test data: \n",confusion_matrix(y_test,y_test_pred))
```

```python
from sklearn.metrics import classification_report
print("Classification report on test data: \n",classification_report(y_test,y_test_pred))
```

The model is a bit over-trained: accuracy on train data is almost 100%, accuracy on test data is 95.21%. The model detected almost all of the cases in which income was below 50K and 63% of the cases in which income is above 50K. 

Oversampling the minority class increased the model's sensitivity to the minority class. However, because the oversampling was done through direct replication, the result is a slightly over-trained model. I think that if synthetic data was used, the result could have been more satisfactory, but it is difficult to generate realistic synthetic census data.

```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_ed1d1e1872004faf87fa9b69405df7ef.csv"
df=pd.read_csv(url,header=None,na_values=["?"])
df.head()
```

```python
df.shape
```

```python
df=df.rename(columns={0:"Class"})
df.groupby("Class")["Class"].count()
```

```python
df.isna().any()
```

```python
df.dtypes
```

```python
df=df.fillna(df.median())
df.isna().any()
```

```python
df1=df[df["Class"]=="2-4-d-injury"]
df1
```

```python
df2=df[df["Class"]=="herbicide-injury"]
df2
```

```python
sub_df=pd.concat([df1,df2],axis=0)
df=df.drop(index=sub_df.index)
df.shape
```

```python
y=df[["Class"]]
X=df.drop(["Class"],axis=1)
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=mutual_info_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
X_scores=scores[scores["scores"]>=0.5]
len(X_scores)
```

```python
X=X.iloc[:,X_scores.index]
X.head()
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
import numpy as np
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
BForest=BalancedRandomForestClassifier(n_estimators=200,criterion="entropy",max_depth=7,min_samples_split=3,
                                       min_samples_leaf=1,max_features="sqrt",bootstrap=False,
                                       sampling_strategy="auto",replacement=True,n_jobs=-1,warm_start=False,
                                       random_state=0)
scores=cross_val_score(BForest,X,np.asarray(y).reshape(y.shape[0],),cv=12)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.naive_bayes import GaussianNB
BBagging=BalancedBaggingClassifier(base_estimator=GaussianNB(),n_estimators=26,sampling_strategy="all",n_jobs=-1,
                                   random_state=0)
scores=cross_val_score(BBagging,X,np.asarray(y).reshape(y.shape[0],),cv=10)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_ed1d1e1872004faf87fa9b69405df7ef.csv"
df=pd.read_csv(url,header=None,na_values=["?"])
```

```python
df=df.fillna(df.median())
```

```python
df=df.rename(columns={0:"Class"})
```

```python
X=df.drop(["Class"],axis=1)
y=df[["Class"]]
```

```python
test = SelectKBest(score_func=mutual_info_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

I will use 13 predictor variables as well.

```python
X_scores=scores[scores["scores"]>0.42]
len(X_scores)
```

```python
X=X.iloc[:,X_scores.index]
```

```python
BForest=BalancedRandomForestClassifier(n_estimators=200,criterion="entropy",max_depth=7,min_samples_split=3,
                                       min_samples_leaf=1,max_features="sqrt",bootstrap=False,
                                       sampling_strategy="auto",replacement=True,n_jobs=-1,warm_start=False,
                                       random_state=0)
scores=cross_val_score(BForest,X,np.asarray(y).reshape(y.shape[0],),cv=12)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

As shown, if the two "outlier" classes were not dropped, with the same number of predictor variables and the same hyper-parameter values, the BalancedRandomForest model's performance is about 17% lower. By simply dropping those two outlier classes, I improved the model's performance by 17%.

```python
sub_y=sub_df[["Class"]]
```

```python
sub_X.shape
```

```python
test = SelectKBest(score_func=mutual_info_classif, k="all")
fit = test.fit(sub_X, np.asarray(sub_y).reshape(sub_y.shape[0],))
features = fit.transform(sub_X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
sub_scores=scores[scores["scores"]>=1.5]
len(sub_scores)
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
clusterer = KMeans(n_clusters=2, random_state=0)
cluster_labels = clusterer.fit_predict(sub_X)
```

```python
label=pd.DataFrame(cluster_labels).rename(columns={0:"Label"})
```

```python
sub_X=sub_X.reset_index(drop=True)
sub_y=sub_y.reset_index(drop=True)
```

```python
outliers=pd.concat([sub_X,sub_y,label],axis=1)
outliers
```

KMeans clustering with K=2 perfectly labeled the two outlier classes.

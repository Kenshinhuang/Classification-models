```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_cf5705cec3fe432b93985616c8ad4532.csv"
df=pd.read_csv(url,header=None,na_values=["?"])
df.head()
```

```python
df.shape
```

```python
df=df.fillna(df.median())
```

```python
df.groupby(df.iloc[:,49]).count().T.head()
```

```python
X=df.iloc[:,0:49]
X.head()
```

```python
y=pd.DataFrame(df.iloc[:,49]).rename(columns={49:"Class"})
y.head()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
X_scores=scores[scores["scores"]>10]
len(X_scores)
```

```python
X=X.iloc[:,X_scores.index]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X))
X.head()
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
import numpy as np
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
BForest=BalancedRandomForestClassifier(n_estimators=180,criterion="entropy",max_depth=5,min_samples_split=4,min_samples_leaf=0.006,max_features="sqrt",bootstrap=True,sampling_strategy="not minority",replacement=False,n_jobs=-1,warm_start=False,random_state=0)
scores=cross_val_score(BForest,X,np.asarray(y).reshape(y.shape[0],),cv=8)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
y_pred = cross_val_predict(BForest, X, np.asarray(y).reshape(y.shape[0],), cv=8)
AUC = roc_auc_score(y, y_pred)
print ("AUC: ", AUC)
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("Specificity: ",51/(51+12))
print("Sensitivity: ",77/(77+25))
```

```python
y_pred=pd.DataFrame(cross_val_predict(BForest, X, np.asarray(y).reshape(y.shape[0],), cv=8,method='predict_proba')).rename(columns={0:"pred_0",1:"pred_1"})
y_pred
```

```python
y_probs1=y_pred[["pred_1"]]
```

```python
import numpy as np
thresholds=np.arange(0,1,0.001)
```

```python
def to_labels(pos_probs,threshold):
    return (pos_probs >= threshold).astype('int')
```

```python
from sklearn.metrics import f1_score
scores=[f1_score(y, to_labels(y_probs1,t),average="micro") for t in thresholds]
```

```python
ix=np.argmax(scores)
print((thresholds[ix],scores[ix]))
```

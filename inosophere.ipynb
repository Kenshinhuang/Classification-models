```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_c69978cffe0240e4adf5f078bca88f8f.csv"
df=pd.read_csv(url,header=None)
df.head()
```

```python
df.shape
```

```python
df=df.rename(columns={34:'Class'})

y=df[["Class"]]

y.groupby("Class")["Class"].count()
```

```python
X=df.drop(["Class"],axis=1)
X.head()
```

```python
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=f_classif, k="all")
fit = test.fit(X, np.asarray(y).reshape(y.shape[0],))
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores=scores.sort_values(by=["scores"],ascending=False)
scores
```

```python
scores=scores.fillna(0)
```

```python
X_scores=scores[scores["scores"]>13]
len(X_scores)
```

```python
Xp=X.iloc[:,X_scores.index]
Xp.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Xp,y,test_size=0.3,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
y_train.groupby("Class")["Class"].count()
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)
```

```python
print(X_res.shape)
```

```python
y_res.groupby("Class")["Class"].count()
```

```python
from sklearn.ensemble import RandomForestClassifier
Forest=RandomForestClassifier(random_state=0).fit(X_res,np.asarray(y_res).reshape(y_res.shape[0],))
Forest
```

```python
from sklearn.metrics import accuracy_score
train_pred=Forest.predict(X_res)
test_pred=Forest.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_res,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
Forest2=RandomForestClassifier(random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
train_pred=Forest2.predict(X_train)
test_pred=Forest2.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
0.9622641509433962 -  0.9433962264150944
```

#### For RandomForestClassifier, resampling improved test data accuracy by 1.9%.

```python
from sklearn.ensemble import GradientBoostingClassifier
Gradient=GradientBoostingClassifier(
                                    random_state=0).fit(X_res,np.asarray(y_res).reshape(y_res.shape[0],))
Gradient
```

```python
train_pred=Gradient.predict(X_res)
test_pred=Gradient.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_res,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

```python
Gradient2=GradientBoostingClassifier(
                                    random_state=0).fit(X_train,np.asarray(y_train).reshape(y_train.shape[0],))
train_pred=Gradient2.predict(X_train)
test_pred=Gradient2.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,train_pred))
print("Accuracy on test data: ",accuracy_score(y_test,test_pred))
```

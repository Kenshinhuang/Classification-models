```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_cf5705cec3fe432b93985616c8ad4532.csv"
df=pd.read_csv(url,header=None,na_values=["?"])
df.head()
```

```python
df=df.fillna(df.median())
```

```python
X=df.iloc[:,0:49]
X.head()
```

```python
y=pd.DataFrame(df.iloc[:,49]).rename(columns={49:"Class"})
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X))
X.head()
```

```python
from sklearn.cluster import KMeans
clusterer = KMeans(n_clusters=2, random_state=0)
cluster_labels = pd.DataFrame(clusterer.fit_predict(X)).rename(columns={0:'Label'})
cluster_labels
```

```python
df=pd.concat([X,y,cluster_labels],axis=1)
df.head()
```

```python
X=df.drop(["Class"],axis=1)
```

```python
df.groupby("Label").count().T.head()
```

```python
y=df[["Class"]]
```

```python
Label=df[["Label"]]
```

```python
y.to_csv("Class.csv")
Label.to_csv("Label.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
train=pd.concat([X_train,y_train],axis=1)
train0=train[train["Label"]==0]
train0=pd.concat([train0]*2,axis=0)
train1=train[train["Label"]==1]
train=pd.concat([train0,train1],axis=0)
train.shape
```

```python
X_train=train.drop(["Label","Class"],axis=1)
X_train.head()
```

```python
X_train=X_train.iloc[:,[26,40,31,35,48,28,21]]
X_train.head()
```

```python
y_train=train[["Class"]]
```

```python
X_test=X_test.drop(["Label"],axis=1)
X_test=X_test.iloc[:,[26,40,31,35,48,28,21]]
X_test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
Forest=RandomForestClassifier(n_estimators=200,criterion="entropy",max_depth=1,ccp_alpha=0.01,random_state=42).fit(X_train,y_train)
Forest
```

```python
from sklearn.metrics import accuracy_score
ytrain_pred=Forest.predict(X_train)
ytest_pred=Forest.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,ytrain_pred))
print("Accuracy on test data: ",accuracy_score(y_test,ytest_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
Gradient=GradientBoostingClassifier(loss="exponential",learning_rate=0.07,criterion="mse",max_depth=1,max_features="sqrt",random_state=42).fit(X_train,y_train)
Gradient
```

```python
ytrain_pred=Gradient.predict(X_train)
ytest_pred=Gradient.predict(X_test)
print("Accuracy on train data: ",accuracy_score(y_train,ytrain_pred))
print("Accuracy on test data: ",accuracy_score(y_test,ytest_pred))
```

```python
from sklearn.metrics import roc_auc_score
print("AUC on train data: ",roc_auc_score(y_train,ytrain_pred))
print("AUC on test data: ",roc_auc_score(y_test,ytest_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix: \n",confusion_matrix(y_test,ytest_pred))
```

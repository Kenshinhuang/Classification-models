Load the data.

```python
import pandas as pd
df=pd.read_csv("Parkinson_dataset.csv")
df
```

The task associated with this dataset is to predict the presence of Parkinson's disease. Let's see how many positive cases, i.e., patients with Parkinson's, there are in the datast.

```python
len(df[df["status"]==1])
```

There are 147 positive cases. The entire dataset has 195 instances. So, there are 48 negative cases (patients with no Parkinson's).

The entire dataset has 24 columns. Excluding the target column "status", there are 23 potential predictor variables. The "name" column has no predictive value. So, we are going to drop that column. There are still 22 potential predictor variables. Some of them may not be relevant. So, we are going to run a feature selection algorithm and compute the feature importance scores for all 22 predictor variables. 

```python
df=df.drop(["name"],axis=1)
X=df.drop(["status"],axis=1)
X.head()
```

```python
y=df[["status"]]
y.head()
```

Compute the feature importance scores for each of the 22 predictor variables.

```python
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

# feature extraction. We are going to output the selection scores for all features and select the features with the highest scores.
test = SelectKBest(score_func=mutual_info_classif, k="all")
fit = test.fit(X, y)
features = fit.transform(X)
# summarize scores
set_printoptions(precision=3)

scores=fit.scores_
scores=pd.DataFrame(scores)
scores=scores.rename(columns={0:"scores"})
scores.sort_values(by=["scores"],ascending=True)
```

All features are relevant. We are going to use 10 of them. So, let's modify the code block above and output only the 10 features with the highest feature importance scores.

```python
scores.sort_values(by=["scores"],ascending=True).nlargest(10,"scores")
```

So, our predictor set is:

```python
X=X.iloc[:,[14,15,1,2,4,12,19,0,18,21]]
X
```

For this dataset, I am not going to implement any outlier removal algorithm. This is a small dataset. I need all the data to do a meaningful analysis and train the model.

Let's standardize the predictor variables.

```python
from sklearn import preprocessing
names=X.columns
scaler=preprocessing.StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X, columns=names)
X.head()
```

One-hot encode the target variable.

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder(handle_unknown="ignore")
y=pd.DataFrame(enc.fit_transform(y).toarray())
y.head()
```

Split the predictor variables and the target variable into train test subsets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15,random_state=42)
```

Build the neural network model, train it and test it.

```python
length=X_train.shape[1]
num_classes=y_test.shape[1]

# Load the libraries
import keras
from keras.models import Sequential
from keras.layers import Dense

# define classification model

def classification_model():
    # create model
    model = Sequential()
    model.add(Dense(length, activation='relu', input_shape=(length,)))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation='relu'))
    model.add(Dense(50, activation= 'relu'))
    model.add(Dense(50, activation = 'relu'))
    model.add(Dense(num_classes, activation='softmax'))
    
    
    # compile model
    model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model=classification_model()
model.fit(X_train,y_train,validation_data=(X_test, y_test),epochs=18, verbose=2)
model.evaluate(X_test,y_test,verbose=0)
```

The neural network model has a 93.33% accuracy on test data. That is pretty good.

Compare the actual labels and predicted labels and compute the confusion a matrix.

The neural network model above is a multilabel classification model. The confusion_matrix function from "from sklearn.metrics import confusion_matrix" could not handle a multilable output. And I got an import error every time when I ran "from sklearn.metrics import multilable_confusion_matrix", for which I had not found a workable solution. So, I am going to mannually compute the confusion matrix. That would be a good exercise.

Join the actual labels and predicted labels.

```python
import numpy as np
y_pred=pd.DataFrame(model.predict(X_test)).rename(columns={0:"pred_0",1:"pred_1"})

y_test=pd.DataFrame(np.asarray(y_test)).rename(columns={0:"actual_0",1:"actual_1"})

label=pd.concat([y_test,y_pred],axis=1)

label.head()
```

We see that the predicted outputs of the model are nothing like the actual binary labels. That is because the model computed the probability that an instance would belong to a particular class. For first instance, the probability of it belonging to Class "0" (no Parkinson's) is 0.000002, or 0.0002%, while the probability of it belonging to Class "1" (Parkinson's) is 0.999998, or 99.9998%. So, the first instance belongs to Class 1. Using a threshold of 0.5, probabilities below 50% will be classified as non-presence of that class. 

```python
import numpy as np
pred_1=np.where(label["pred_1"]>=0.5)
print("The row indices of the predicted positive cases: ",pred_1)
print("The number of positive cases predicted by the model: ", len(np.asarray(pred_1).T))
```

The model predicted 28 positive cases. Let's see how many actual cases there are in the test data.

```python
print("There are ",len(label[label["actual_1"]==1])," actual positive cases in the test data.")
```

Let's pick out the actual positive cases and see how that compare to the predicted positive cases. We want to see the model's sensitivity. How well did the model performed on the predicting the cases for which the patients did have Parksinson's disease?

```python
sub_label_1=label[label["actual_1"]==1]
missed_sub_label_1=sub_label_1[sub_label_1["pred_1"]<0.5]
missed_sub_label_1
```

The first line of the above code block picked out the actual positive cases, formed a new dataframe with only instances. Then, the second line examined whether there  is any case in this new dataframe that the model had predicted to be negative. The code block output nothing. That means the model correctly predicted ALL positive cases. It correclty predicted all the instances that the patients had Parkinson's disease. There are a total of 26 positive cases in the test data. The model predicted 28. So, True Positive is 26 and False Positive is 2.

Let's also see how the model performed on the negative cases in the test data.

```python
pred_0=np.where(label["pred_0"]>=0.5)
pred_0
```

The model only predicted 2 negative cases. Let's see how that compared to the actual number of negative cases in the test data.

```python
sub_label_0=label[label["actual_0"]==1]
sub_label_0
```

There are 4 negative cases in the test data. The model predicted 2. Hence, True Negative is 2. 

We got our confusion matrix.

```python
print("Confusion matrix: \n", np.array([[26,0],[2,2]]))
```

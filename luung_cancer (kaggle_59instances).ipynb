```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_9a0760878dc7470aaa032c2cfbce892b.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
y=df[["Result"]]
y.value_counts()
```

```python
X=df[["Age","Smokes","AreaQ","Alkhol"]]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
X.head()
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
```

```python
from sklearn import svm
import numpy as np
model1=svm.SVC(C=2,random_state=0)
scores=cross_val_score(model1,X,np.asarray(y).reshape(y.shape[0],),cv=3)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

A 2-fold cross validation would give the same average accuracy. Standard deviation with 2-fold is 1.7%, 0.7% lower than 3-fold.

```python
scores
```

```python
y_pred = cross_val_predict(model1, X, np.asarray(y).reshape(y.shape[0],), cv=3)
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("Classification report: \n",classification_report(y,y_pred))
```

The model missed only 1 instance.

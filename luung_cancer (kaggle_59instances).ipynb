```python
import pandas as pd
url="https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_9a0760878dc7470aaa032c2cfbce892b.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
y=df[["Result"]]
y.value_counts()
```

```python
X=df[["Age","Smokes","AreaQ","Alkhol"]]
X.head()
```

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
```

```python
from sklearn import svm
import numpy as np
model1=svm.SVC(C=5,kernel="linear",random_state=0)
scores=cross_val_score(model1,X,np.asarray(y).reshape(y.shape[0],),cv=2)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

```python
y_pred = cross_val_predict(model1, X, np.asarray(y).reshape(y.shape[0],), cv=2)
print("Confusion matrix: \n",confusion_matrix(y,y_pred))
```

```python
print("Classification report: \n",classification_report(y,y_pred))
```

```python
X=df[["Smokes"]]
```

```python
model2=svm.SVC(C=4,kernel="poly",gamma="auto",degree=2,random_state=0)
scores=cross_val_score(model2,X,np.asarray(y).reshape(y.shape[0],),cv=2)
print("Accuracy: ",scores.mean(),"+/-",scores.std())
```

#### Smoking is by far the most important factor in predicting lung cancer. Using only "Smokes" as input, we could obtain a model that is 73% correct. In other words, just tell me whether someone smokes or not, I can tell whether that person will develop lung cancer or not and 73% of the time, I will be correct. 

#### If we include other factors such as age and alcohol usage, the model is 95% correct.
